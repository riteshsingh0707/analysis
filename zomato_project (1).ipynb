```python
#handling all missing data
# Drop columns not needed
df.drop(columns=['zipcode','opentable_support'],inplace=True)
```

```python
# Fill missing 'timings' and 'address'
df['timings'].fillna('not Available',inplace=True)
df['address'].fillna('not Available',inplace=True)
```

```python
# Fill missing 'cuisines' with 'Unknown'
df['cuisines'].fillna('Unknown', inplace=True)
```

```python
# Recheck missing data
df.isnull().sum().sort_values(ascending=False)
```

```python
#Distribution of Aggregate Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['aggregate_rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='rating_text', order=df['rating_text'].value_counts().nlargest(10).index, palette='Set2')
plt.title("Count of Restaurants by Rating Category")
plt.xlabel("Rating Text")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
average_rating = df['aggregate_rating'].mean()
print("Average Rating of All Restaurants:", round(average_rating, 2))
```

```python
# Count top 10 cuisines
cuisine_counts=df['cuisines'].value_counts().head(10)
print("Top 10 Cuisines:")
print(cuisine_counts)
```

```python
# Plot top cuisines
plt.figure(figsize=(10, 5))
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index, palette='magma')
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.show()
```

```python
#Correlation Between Cuisine Variety & Rating
# Count number of cuisines per restaurant
df['cuisine_variety'] = df['cuisines'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
# Correlation coefficient
correlation = df['cuisine_variety'].corr(df['aggregate_rating'])
print("Correlation between cuisine variety and rating:", round(correlation, 3))
```

```python
# Boxplot: variety vs rating
plt.figure(figsize=(8, 5))
sns.boxplot(x='cuisine_variety', y='aggregate_rating', data=df)
plt.title("Cuisine Variety vs Aggregate Rating")
plt.xlabel("Number of Cuisines Offered")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.show()
```

```python
#Price Range vs Rating
plt.figure(figsize=(8, 5))
sns.boxplot(x='price_range', y='aggregate_rating', data=df, palette='pastel')
plt.title("Price Range vs Aggregate Rating")
plt.xlabel("Price Range (1 = Low, 4 = High)")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.show()
```

```python
# Grouping average cost by price range
avg_cost_by_price = df.groupby('price_range')['average_cost_for_two'].mean()

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_cost_by_price.index, y=avg_cost_by_price.values, palette='coolwarm')
plt.title("Average Cost for Two in Each Price Range")
plt.xlabel("Price Range (1 = Low, 4 = High)")
plt.ylabel("Average Cost for Two (‚Çπ)")
plt.grid(True)
plt.show()
```

```python
correlation = df['price_range'].corr(df['aggregate_rating'])
print("Correlation between price range and rating:", round(correlation, 3))
```

```python
 #Online Order and Table Booking Analysis
#Assuming
#delivery = 1 means Online Order is available
#delivery = -1 means Not available
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.boxplot(x='delivery', y='aggregate_rating', data=df, palette='Set2')
plt.title("Online Order Availability vs Ratings")
plt.xlabel("Online Order (1 = Yes, -1 = No)")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.show()
```

```python
# Distribution of Restaurants with Table Booking
#We‚Äôll use the highlights column to detect whether "Table booking" is mentioned.
# Create a new column indicating if table booking is available
df['has_table_booking'] = df['highlights'].apply(lambda x: 'Table booking' in x)

# Count of restaurants offering table booking
table_booking_counts = df['has_table_booking'].value_counts()

# Bar plot of distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='has_table_booking', data=df, palette='pastel')
plt.title("Distribution of Restaurants Offering Table Booking")
plt.xlabel("Table Booking Available")
plt.ylabel("Number of Restaurants")
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
#Impact of Table Booking on Ratings
plt.figure(figsize=(6, 4))
sns.boxplot(x='has_table_booking', y='aggregate_rating', data=df, palette='coolwarm')
plt.title("Table Booking Availability vs Ratings")
plt.xlabel("Table Booking Available")
plt.ylabel("Aggregate Rating")
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
 #Identify Top Restaurant Chains by Number of Outlets
 # Count number of outlets per restaurant name
top_chains = df['name'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_chains.values, y=top_chains.index, palette='cubehelix')
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Name")
plt.show()
```

```python
#Explore Ratings of These Top Chains
# Filter data for only top 10 restaurant chains
top_chain_names = top_chains.index.tolist()
top_chain_data = df[df['name'].isin(top_chain_names)]

# Plot average rating of each top chain
chain_avg_ratings = top_chain_data.groupby('name')['aggregate_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(x=chain_avg_ratings.values, y=chain_avg_ratings.index, palette='coolwarm')
plt.title("Average Rating of Top 10 Restaurant Chains")
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Name")
plt.show()
```

```python
# Optional: Show number of votes for each chain
chain_votes = top_chain_data.groupby('name')['votes'].sum().sort_values(ascending=False)
print("Total Votes per Chain:\n", chain_votes)
```

```python
flat_highlights = (df['highlights'].dropna().str.split(',').explode().str.strip())
top_features = flat_highlights.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 4))
sns.barplot(x=top_features.values, y=top_features.index, palette='Blues_d')
plt.title("Top 10 Most Common Restaurant Features")
plt.xlabel("Number of Restaurants")
plt.ylabel("Feature")
plt.show()
```

```python
#Correlation Between a Feature (e.g., Wi-Fi) and Ratings
# Choose a feature to test
feature_to_test = 'Wi-Fi'

# Create a new column indicating if the feature exists
df['has_wifi'] = df['highlights'].apply(lambda x: feature_to_test in x)

# Compare ratings using boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x='has_wifi', y='aggregate_rating', data=df, palette='Set2')
plt.title(f"Restaurant Ratings: With vs Without '{feature_to_test}'")
plt.xlabel(f"'{feature_to_test}' Available")
plt.ylabel("Aggregate Rating")
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
#Word Cloud from Reviews
#Create a Word Cloud from Review Text (Assuming a reviews column exists)
from wordcloud import WordCloud

# Assuming you have a review column:
#reviews_text = df['reviews'].dropna().str.cat(sep=' ')

# Since reviews don't exist, we simulate with rating_text
reviews_text = df['rating_text'].dropna().str.cat(sep=' ')

# Generate Word Cloud
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='viridis').generate(reviews_text)

# Plot
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud from Rating Text")
plt.show()
```

```python
#Sentiment Classification Based on rating_text
# Simple sentiment mapping
def classify_sentiment(rating_text):
    if rating_text in ['Excellent', 'Very Good']:
        return 'Positive'
    elif rating_text in ['Average']:
        return 'Neutral'
    else:
        return 'Negative'

df['sentiment'] = df['rating_text'].apply(classify_sentiment)
#ploting graph
plt.figure(figsize=(6, 4))
sns.countplot(x='sentiment', data=df, palette='Set1')
plt.title("Customer Sentiment Based on Ratings")
plt.xlabel("Sentiment")
plt.ylabel("Number of Restaurants")
plt.grid(True)
plt.show()
```
